```python
# Decode data: 

from datetime import datetime, timedelta

def create_timestamp(i):
    date = datetime.strptime(i['OPD_DATE'],'%d%b%Y:%H:%M:%S')
    time = timedelta(seconds=i['ACT_TIME'])
    return date + time 
    
df['TIMESTAMP'] = df.apply(create_timestamp,axis=1)

date = datetime.strptime('15FEB2023:00:00:00', '%d%b%Y:%H:%M:%S')
time = timedelta(seconds=20469)

    
```

```python
df = df.drop(columns=['OPD_DATE', 'ACT_TIME'])
df
```

```python
df['dMETERS'] = df['METERS'].diff()
df['dTIMESTAMP'] = df['TIMESTAMP'].diff()

df
```

```python
df = df.drop(labels=0, axis=0)
df
```

```python
def speed(meters, time):
    speed = meters/time
    return speed

df['SPEED'] = df.apply(lambda row: speed(row['dMETERS'],row['dTIMESTAMP'].total_seconds()), axis=1)
df
```

```python
df = df.drop(columns=['dMETERS','dTIMESTAMP'])
df
```

```python
# from datetime import datetime, timedelta

def create_timestamp(i):
    date = datetime.strptime(i['OPD_DATE'],'%d%b%Y:%H:%M:%S')
    time = timedelta(seconds=i['ACT_TIME'])
    return date + time 
    
df2['TIMESTAMP'] = df2.apply(create_timestamp,axis=1)

date = datetime.strptime('15FEB2023:00:00:00', '%d%b%Y:%H:%M:%S')
time = timedelta(seconds=20469)
df2 = df2.drop(columns=['OPD_DATE', 'ACT_TIME'])

df2.head(5)
```

```python
df2 = df2.sort_values(by=['EVENT_NO_TRIP', 'TIMESTAMP'])
```

```python
df2['dMETERS'] = df2['METERS'].diff()
df2['dTIMESTAMP'] = df2['TIMESTAMP'].diff()
df2.head(5)
```

```python
df2 = df2.drop(labels=[3617,2313],axis=0)
df2.head(5)
```

```python
def speed(meters, time):
    speed = meters/time
    return speed

df2['SPEED'] = df2.apply(lambda row: speed(row['dMETERS'],row['dTIMESTAMP'].total_seconds()), axis=1)
df2 = df2.drop(columns=['dMETERS','dTIMESTAMP'])
df2
```
