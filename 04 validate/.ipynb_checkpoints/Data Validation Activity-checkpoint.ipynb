```python
import pandas as pd

data = pd.read_csv("Hwy26Crashes2019_S23.csv")

data
```

```python
"""Assertion that all crash IDs have a record type field (existence)"""
def crashID_has_recordType(): 
    for i in range(2739):
        if data.iloc[i]['Crash ID'] > 0 and data.iloc[i]['Record Type'] > 0:
            return True

crashID_has_recordType()
```
